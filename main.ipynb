```python
def sec_to_time(sec):
    minute = str(sec // 60)
    sec = str(sec % 60)
    if len(minute) == 1:
        minute = "0" + minute
    if len(sec) == 1:
        sec = "0" + sec
    return f'00:{minute}:{sec},000'
```

```python
txt = """"""
for i in range(168):
    new_txt = \
f"""{i + 1}
{sec_to_time(i)} --> {sec_to_time(i+1)}
{i}

"""
    txt += new_txt 
```

```python
with open("sample_video.srt", "w") as f:
    f.write(txt)
```
